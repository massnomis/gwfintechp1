```python
class CurrencyConfig:
    def __init__(self, Currency_Name, MaximumLTV, Liquidationthreshold, Liquidationpenalty, Ideal_UTIL, BASE, at_ideal, at_Full, Reserve_Factor):
       self.Currency_Name = Currency_Name 
       self.MaximumLTV = MaximumLTV 
       self.Liquidationthreshold = Liquidationthreshold
       self.Liquidationpenalty = Liquidationpenalty
       self.Ideal_UTIL = Ideal_UTIL
       self.BASE = BASE
       self.at_ideal = at_ideal
       self.at_Full = at_Full
       self.Reserve_Factor = Reserve_Factor
       
    def UTIL (self, Borrows, Total_Liquidity):
        return (Borrows / Total_Liquidity)
             
        
    def Variable_Rate (self, Borrows, Total_Liquidity):
        if self.UTIL(Borrows, Total_Liquidity) < self.Ideal_UTIL:
            return (self.UTIL(Borrows, Total_Liquidity) / self.Ideal_UTIL) * self.at_ideal
        elif self.UTIL(Borrows, Total_Liquidity) >= self.Ideal_UTIL:
            return self.at_ideal + ( self.UTIL(Borrows, Total_Liquidity) - self.Ideal_UTIL /  (1 - self.Ideal_UTIL )) * self.at_Full

        
    def Lend_APY (self, Borrows, Total_Liquidity):
        return (self.Variable_Rate(Borrows, Total_Liquidity)) * (1 - self.Reserve_Factor)


    
    
    
MATIC = CurrencyConfig("MATIC", 0.5, 0.65, 0.1, 0.45, 0, 0.07, 3, 0.2)

MATICBorrows = 1
MATICTotal_Liquidity = 100

print(MATIC.Lend_APY(Borrows, Total_Liquidity))
print(MATIC.Ideal_UTIL)
```
